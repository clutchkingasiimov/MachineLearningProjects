```python
# Extract feature columns
feature_cols = list(df.columns[:-1])

# Extract target column 'passed'
target_col = df.columns[-1] 

# Show the list of columns
print("Feature columns:\n{}".format(feature_cols))
print("\nTarget column: {}".format(target_col))

# Separate the data into feature data and target data (X_all and y_all, respectively)
X_all = df[feature_cols]
y_all = df[target_col]

# Show the feature information by printing the first five rows
print("\nFeature values:")
print(X_all.head())
```

### Preprocess Feature Columns

The data contains a lot of non-numeric values which need to be converted which can be made into boolean values

Categories like Mjob and Fjob have more than 2 values and are known as categorical variables. The recommended way to handle such a column is to create as many columns as possible values (e.g. Fjob_teacher, Fjob_other, Fjob_services, etc.), and assign a 1 to one of them and 0 to all others.

```python
def preprocess_features(x): #Preprocesses the student data and converts non-numeric binary to binary and categorical to dummy variables.
    
    #Initialize new output dataframe
    output = pd.DataFrame(index = x.index)
    
    #Investigate each feature column for the data
    for col, col_data in x.iteritems():#.iteritems iterates over pairs
        
        #If data type is non-numeric, replace all yes/no values with 1/0
        if col_data.dtype == object:
            col_data = col_data.replace(['yes', 'no'], [1, 0])
            
        #If data type is categorical, convert to dummy variables
        if col_data.dtype == object:
            col_data = pd.get_dummies(col_data, prefix = col)
        
        #Collect the updated columns
        output = output.join(col_data)
    
    return output

X_all = preprocess_features(X_all)
X_all.head(10)
                                         
  
```

```python
# First, decide how many training vs test samples you want
num_all = df.shape[0]  # same as len(student_data)
num_train = 300  # about 75% of the data
num_test = num_all - num_train

# TODO: Then, select features (X) and corresponding labels (y) for the training and test sets
# Note: Shuffle the data or randomly select samples to avoid any bias due to ordering in the dataset
#student_data['passed'] = format(student_data['passed'])
y = df['passed']

def stratify_and_shuffle(X,y,num_train):
    sss = StratifiedShuffleSplit(y, 1, train_size=num_train)

    # only one iteration
    for train_index, test_index in sss:
        X_train = X.iloc[train_index]
        X_test = X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
    return X_train, X_test, y_train, y_test
    
X_train, X_test, y_train, y_test = stratify_and_shuffle(X_all, y, num_train)
    
print("Training set: {} samples".format(X_train.shape[0]))
print("Test set: {} samples".format(X_test.shape[0]))
```

# Training and Evaluating Models

A table will be produced showing training prediction time, F1 score on training set and F1 score on testing set for each training size

```python
# Train a model
import time

def train_classifier(clf, X_train, y_train):
    print("Training {}...".format(clf.__class__.__name__))
    start = time.time()
    clf.fit(X_train, y_train)
    end = time.time()
    diff = end - start
    print("Done!\nTraining time (secs): {:.3f}".format(diff))
    return diff

# TODO: Choose a model, import it and instantiate an object

from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=30)

# Fit model to training data
train_classifier(rf_clf, X_train, y_train)  # note: using entire training set here
print(rf_clf)  # you can inspect the learned model by printing it
```

```python
#Predict on training set and compute the F1 score
from sklearn.metrics import f1_score

def predict_labels(clf, features, target):
    print('Label prediction using {}...'.format(clf.__class__.__name__))
    start = time.time()
    y_pred = clf.predict(features)
    end = time.time()
    prediction_time_diff = end - start
    print('Procecss complete \nPrediction time (secs): {:.3f}'.format(prediction_time_diff))
    return (f1_score(target.values, y_pred, pos_label = 'yes'), prediction_time_diff)

train_f1_score = predict_labels(rf_clf, X_train, y_train)
print('F1 score for training set: {}'.format(train_f1_score[0]))
```
