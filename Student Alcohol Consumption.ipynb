### Confidence interval for a sample size of 100 from the dataset

```python
#n = 100
#x-bar = 10.41
#sigma = 4.58
import math
def ci_range(xbar, z, sigma, n, symbol):
    if symbol == 'plus':
        range = xbar + z * (sigma / math.sqrt(n))
        return range
    
    elif symbol == 'minus':
        range = xbar - z * (sigma / math.sqrt(n))
        return range
    
upperrange = ci_range(10.41, 1.96, 4.58, 395, 'plus')
lowerrange = ci_range(10.41, 1.96, 4.58, 395, 'minus')
lowerrange, upperrange

#We are sure that 95% of our data used for finding the true mean lies between 10.12 and 11.4
```

## 1. Data Extraction
### Identify feature and target columns

Often the case is that the data can have non-numeric features which can be a problem for the algorithms to perform the coding process

```python
# Extract feature columns
feature_cols = list(df.columns[:-1])

# Extract target column 'passed'
target_col = df.columns[-1] 

# Show the list of columns
print("Feature columns:\n{}".format(feature_cols))
print("\nTarget column: {}".format(target_col))

# Separate the data into feature data and target data (X_all and y_all, respectively)
X_all = df[feature_cols]
y_all = df[target_col]

# Show the feature information by printing the first five rows
print("\nFeature values:")
print(X_all.head())
```

### Preprocess Feature Columns

The data contains a lot of non-numeric values which need to be converted which can be made into boolean values

Categories like Mjob and Fjob have more than 2 values and are known as categorical variables. The recommended way to handle such a column is to create as many columns as possible values (e.g. Fjob_teacher, Fjob_other, Fjob_services, etc.), and assign a 1 to one of them and 0 to all others.

```python
def preprocess_features(x): #Preprocesses the student data and converts non-numeric binary to binary and categorical to dummy variables.
    
    #Initialize new output dataframe
    output = pd.DataFrame(index = x.index)
    
    #Investigate each feature column for the data
    for col, col_data in x.iteritems():#.iteritems iterates over pairs
        
        #If data type is non-numeric, replace all yes/no values with 1/0
        if col_data.dtype == object:
            col_data = col_data.replace(['yes', 'no'], [1, 0])
            
        #If data type is categorical, convert to dummy variables
        if col_data.dtype == object:
            col_data = pd.get_dummies(col_data, prefix = col)
        
        #Collect the updated columns
        output = output.join(col_data)
    
    return output

X_all = preprocess_features(X_all)
X_all.head(10)
                                         
  
```

# Train/Test split



```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
```

```python
#Splitting predicting variables into 'y' and 'X'
y = df['G3']
X = df.drop(['G3'], axis = 1)
```

```python
#Getting dummy variables for binary classification
X = pd.get_dummies(X)
```

```python
codification = ['DecisionTreeRegressor', 'LinearRegression', 'Ridge', 'Lasso']
clf_list = [DecisionTreeRegressor(),
           LinearRegression(),
           Ridge(),
           Lasso()]
```

```python
for code, clf in zip(codification, clf_list):
    print(code, end=': ')
    print(cross_val_score(clf, X, y, cv=5).mean())
```

```python
#Lasso Regression Analysis turned out to give the highest accuracy score of 80.9%
```

```python
#Obtaining the feature importances
tree = DecisionTreeRegressor()
tree.fit(X, y)
```

```python
#Obtaining the predicted values from DecisionTreeRegressor
tree.predict(X)
```

```python
#Coefficient of Determination of the prediction
tree.score(X, y)
```

```python
#Lastly, getting out the feature importances of the algoritm (It was observed that underaverage and second term score played a combined 75% role in predicting the future chances of students consuming alcohol.)
alpha = tree.feature_importances_
indices = np.argsort(alpha)[::-1]
for f in range(X.shape[1]):
    print("%d. Feature %s (%f)" % (f + 1, X.columns.values[indices[f]], alpha[indices[f]]))
```
