```python
import pandas as pd
import numpy as np
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\Sauraj (Work mode)\\Desktop\\temps.csv')
df
```

```python
#Descriptive stats for each column
df.describe()
```

```python
#One-hot encoding the data using pandas to make the machine understand the data better
df = pd.get_dummies(df)
df.head()
```

```python
#Labels are the values we want to predict
labels = np.array(df['actual'])

#Remove the labels from the dataframe and save it as a 'features' variable
features = df.drop('actual', axis = 1)

#Saving feature names for later usage
feature_list = list(features.columns)

#Conversion to array
features = np.array(features)
```

```python
from sklearn.model_selection import train_test_split

#Splitting the data
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
print(train_labels.shape)
```

```python
#Creating a baseline to see if the model can outperform it

baseline_preds = test_features[:, feature_list.index('average')]

#Display baseline errors and display average baseline error
baseline_errors = abs(baseline_preds - test_labels)

print('Average baseline error is :', round(np.mean(baseline_errors), 2))
```

```python
#Model training
from sklearn.ensemble import RandomForestRegressor

#Instantiante the model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)

#Train the model on the training data
rf.fit(train_features, train_labels)
```

```python
#Making predictions on the testing set
predictions = rf.predict(test_features)

#Calculating the error margin
errors = abs(predictions - test_labels)

print('Mean absoulte error of the model:', round(np.mean(errors), 2), 'degrees')
```

```python
#Determining performance metrics
mape = 100 * (errors / test_labels)

#Calculate and display accuracy
accuracy = 100 - np.mean(mape)
accuracy
```

```python
#Variable importances
feature_importances = list(zip(feature_list, rf.feature_importances_))
#Call feature importances according to highest priority
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
feature_importances
```

```python
#Making a random forst with only the two most important variables, temp_1 and averge

rf_important = RandomForestRegressor(n_estimators = 1000, random_state = 42)

#Extraction of two most important features
imp_indices = [feature_list.index('temp_1'), feature_list.index('average')]
train_imp = train_features[:, imp_indices]
test_imp = test_features[:, imp_indices]

#Training the random forest
rf_important.fit(train_imp, train_labels)

#Making predictions to determine the error
predictions = rf_important.predict(test_imp)

errors = abs(predictions - test_labels)

#Displaying the performance metrics
print('MAE:', round(np.mean(errors), 2), 'degrees')

mape = np.mean(100 * (errors / test_labels))
accuracy = 100 - mape
accuracy
```
